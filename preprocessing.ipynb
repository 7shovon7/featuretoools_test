```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data = 'breast-cancer.data'
names = ['Class','age','menopause','tumor-size','inv-nodes','node-caps','deg-malig','breast','breast-quad','irradiat']
dataset = pd.read_csv(data, header=None, names=names)
```

```python
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
```

```python
missing = dataset['deg-malig'].values.reshape(-1,1)
dataset['deg-malig'] = imputer.fit_transform(missing)
```

```python
import featuretools as ft
import numpy as np
import pandas as pd
```

```python
data = ft.demo.load_mock_customer()
```

```python
data
```

```python
transactions = data['transactions']
sessions = data['sessions']
customers = data['customers']
```

```python
transactions.head()
```

```python
sessions.head()
```

```python
customers.head()
```

```python
customers['joined_day'] = customers['join_date'].dt.day
customers['joined_month'] = customers['join_date'].dt.month
customers['joined_year'] = customers['join_date'].dt.year
```

```python
customers.head()
```

```python
transactions_and_sessions = transactions.merge(sessions, on='session_id', how='left')
```

```python
transactions_and_sessions.head()
```

```python
grp = transactions_and_sessions.groupby('customer_id')['amount'].agg(['mean', 'max', 'min'])
grp.columns = ['mean transaction_amount', 'max transaction_amount', 'min transaction_amount']
```

```python
grp.head()
```

```python
ss = customers.merge(grp, on='customer_id', how='left')
```
